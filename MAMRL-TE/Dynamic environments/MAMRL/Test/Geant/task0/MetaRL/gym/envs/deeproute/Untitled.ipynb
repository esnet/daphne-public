```python
import pandas as pd

num_nodes = 12

data = pd.read_csv('b4_demand.txt', header = None)

print(data)
```

```python
float(data[0][0].split()[0])
```

```python
b4_demand = {}
path = 0
for index1 in range(num_nodes):
    b4_demand[index1] = {}
    for index2 in range(num_nodes):
        if index1 == index2:
            continue
        b4_demand[index1][index2] = []
        for index3 in range(len(data[0])):
            demand = data[0][index3].split()[path]
            b4_demand[index1][index2].append(float(demand))
        path += 1
            
            
        
        


```
